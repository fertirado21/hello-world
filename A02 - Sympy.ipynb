**Para la siguientes funciones encuentre los puntos críticos, grafique y diga si éste se trata de un máximo o mínimo, local o global.**

$f(x)=1260+117x-9x^2$

```python
import numpy as np
import sympy as sp
```

```python
sp.var("x")
def f(x):
    return 1260 + 117*x - 9*x**2

df = sp.diff(f(x), x)

pc=sp.solve(df, x)
pc
```

```python
ddf = sp.diff(df, x)
ddf
```

```python
for c in pc:
    crit_2da_derivada = ddf.subs(x, c)
    print(crit_2da_derivada)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
xn = np.linspace(-10, 10, 1000)
plt.figure()
plt.plot(xn, f(xn))
for c in pc:
    plt.plot(c, f(c), "*", ms=10)
```

- El punto crítico esta en 13/2 y es un máximo global

$f(x)=5+x+\frac{4}{x}$, para $x>0$

```python
ddf = sp.diff(df, x)
ddf
```

```python
for c in pc:
    if c > 0:
        crit_2da_derivada = ddf.subs(x, c)
        print(crit_2da_derivada)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
xn = np.linspace(0.1, 10, 1000)
plt.figure()
plt.plot(xn, f(xn))
for c in pc:
    if c > 0:
        plt.plot(c, f(c), "*", ms=10)
```

$f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
sp.var("x")
def f(x):
    return 6 - 2*x + (18-2*x**2)**(1/2)

df = sp.diff(f(x), x)

pc=sp.solve(df, x)
for c in pc:
    if -3 <= c <= 3 :
        print (c)
```

```python
ddf = sp.diff(df, x)
ddf
```

```python
for c in pc:
    if -3 <= c <= 3 :
        crit_2da_derivada = ddf.subs(x, c)
        print(crit_2da_derivada)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
xn = np.linspace(-3, 3, 1000)
plt.figure()
plt.plot(xn, f(xn))
for c in pc:
    if -3 <= c <= 3 :
        plt.plot(c, f(c), "*", ms=10)
```

- El punto crítico está en -2.44 y es un máximo local

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

```python
sp.var("x")
def f(x):
    return (1/4)*x**4 + (1/3)*x**3 -x**2

df = sp.diff(f(x), x)

pc=sp.solve(df, x)
for c in pc:
    if -3 <= c <= 1.7 :
        print (c)
```

```python
ddf = sp.diff(df, x)
ddf
```

```python
for c in pc:
    if -3 <= c <= 1.7 :
        crit_2da_derivada = ddf.subs(x, c)
        print(crit_2da_derivada)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
xn = np.linspace(-3, 1.7, 1000)
plt.figure()
plt.plot(xn, f(xn))
for c in pc:
    if -3 <= c <= 1.7 :
        plt.plot(c, f(c), "*", ms=10)
```

- El primer punto crítico es -2 y es un mínimo global. 
- El segundo punto crítico está en 0 y es un máximo global.
- El tercer punto crítico está en 1 y es un mínimo local.
