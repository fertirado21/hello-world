# Tridilosa

## ¿Qué es y para que sirve la tridilosa?

La **tridilosa** es un invento mexicano el cual revolucionó la industria de la construcción, ya que mejoró la capacidad de las estructuras metálicas para que puedan ser construidas con perfiles ligeros.

## Tipos de tridilosas

1. Metálicas
2. Electrosoldadas

## Ventajas

El uso de tridilosas trae consigo muchas *ventajas*, como por ejemplo:

- Mayor facilidad para poner instalaciones electricas y sanitarias.
- Aislamiento térmico y acústico.
- Funcionan como aligerante por lo que se pueden hacer estructuras menos convencionales.
- Oportunidad de salvar claros de mayor magnitud.
